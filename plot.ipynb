```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy as np

def ema(y, alpha=0.1):
    """Exponential moving average. alpha in (0, 1]; higher = less smoothing."""
    y = np.asarray(y, dtype=float)
    if y.size == 0:
        return y
    out = np.empty_like(y)
    out[0] = y[0]
    for i in range(1, len(y)):
        out[i] = alpha * y[i] + (1 - alpha) * out[i - 1]
    return out
```

```python
folders = [
    "/home/jovyan/avarlamov/.../super-mario-project/logs/reinforce_mario_lr3e-4",
    "/home/jovyan/avarlamov/.../super-mario-project/logs/reinforce_mario_lr1e-4",
    "/home/jovyan/avarlamov/.../super-mario-project/logs/reinforce_mario_lr5e-5",
    "/home/jovyan/avarlamov/.../super-mario-project/logs/reinforce_mario_lr1e-5",

    # "/home/jovyan/avarlamov/.../super-mario-project/logs/mario_lr1e-5",

]

csv_paths = [
    f"{folder}/train_updates.csv" for folder in folders
]

lrs = [
    folder.split("/")[-1].split("_lr")[1] for folder in folders
]

dfs = [pd.read_csv(path).sort_values("update").reset_index(drop=True) for path in csv_paths]
```

```python
plt.figure(figsize=(12, 8))

for df, lr in zip(dfs, lrs):

    df = df[df["update"] <= 5000]

    # plt.plot(df["avg_return"], label=lr)

    # y_smooth = ema(df["mean_ep_x_pos"].values, alpha=1.0)
    plt.plot(df["update"], df["mean_ep_x_pos"], label=f"LR={lr}")

plt.axhline(y=3161, color='black', linestyle='dashed', linewidth=2)
plt.text(
    x=plt.gca().get_xlim()[0] + 1500,   # right edge of the axis
    y=3161,
    s='Finish',
    va='bottom',
    ha='right',
    fontsize=14,
    color='black',
    # backgroundcolor='white'
)

plt.title(f"REINFORCE: X Position During Training (averaged over done rollouts in epoch)", fontsize=16)

plt.legend(fontsize=14)

plt.grid(alpha=0.4)

plt.xlabel("Training Update", fontsize=14)
plt.ylabel("X position", fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
None

plt.savefig("images/reinforce_x_position.png", bbox_inches="tight")
```
